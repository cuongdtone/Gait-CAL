```python
import numpy as np

with open('src/gait_fc_train_result.txt', 'r', encoding='utf8') as f:
    data = f.readlines()
with open('src/gait_fc_train_result_2.txt', 'r', encoding='utf8') as f:
    data2 = f.readlines()

data = data + data2
losses = []
for l in data:
    try:
        i = l.index('Loss=')
        loss = l[i:]
        loss = float(loss.split('=')[-1].strip())
        losses.append(loss)
    except:
        pass
print(losses)
losses = np.array(losses)*20
```

```python
import matplotlib.pyplot as plt

plt.plot(losses)
plt.xlabel('iter')
plt.ylabel('Loss')
```
