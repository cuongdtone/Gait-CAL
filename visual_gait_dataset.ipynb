```python
import glob

import matplotlib.pyplot as plt
import numpy as np
import cv2

images =  glob.glob(r'E:\Gait\dataset\001\nm-02_072\*')
```

```python
def crop(mask):
    x, y, w, h = cv2.boundingRect(mask)
    crop = mask[y:y + h, x:x + w]
    if h > w:
        bg = np.zeros((h, h))
        st_x = int(h / 2 - w / 2)
        ed_x = int(h / 2 + w / 2)
        bg[:, st_x:ed_x] = crop
    else:
        bg = np.zeros((w, w))
        st_y = int(w / 2 - h / 2)
        ed_y = int(w / 2 + h / 2)
        bg[st_y:ed_y, :] = crop
    return bg
```

```python
gei = np.zeros((112, 112), dtype='float')
for idx, img in enumerate(images):
    image = cv2.imread(img, 0)
    mask = crop(image)
    mask = cv2.resize(mask, (112, 112A))
    gei += mask
    # cv2.imwrite(f'samples_gait/{idx}.jpg', mask)
plt.imshow(gei)
```
